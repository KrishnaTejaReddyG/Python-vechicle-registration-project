```
print('Vehicle registration System')
class Automobile:
    def __init__(self):
        self._make = ''
        self._model = ''
        self._year = 0
        self._color = ''
        self._mileage = 0
    def addVehicle(self):
        try:
            self._make = input('Enter vehicle make: ')
            self._model = input('Enter vehicle model: ')
            self._year = int(input('Enter vehicle year: '))
            self._color = input('Enter vehicle color: ')
            self._mileage = int(input('Enter vehicle mileage: '))
            return True
        except ValueError:
            print('Please try entering vehicle information again using only whole numbers for mileage and year')
            return False
    def __str__(self):
        return '\t'.join(str(x) for x in [self._make, self._model, self._year, self._color, self._mileage])

class registration:
    def __init__(self):
        self.vehicles = []
    def addVehicle(self):
        vehicle = Automobile()
        if vehicle.addVehicle() == True:
            self.vehicles.append(vehicle)
            print ()
            print('This vehicle has been added, Thank you')
    def viewregistration(self):
        print('\t'.join(['','Make', 'Model','Year', 'Color', 'Mileage']))
        for idx, vehicle in enumerate(self.vehicles) :
            print(idx + 1, end='\t')
            print(vehicle)

registration = registration()
while True:

    print('#1 Add Vehicle to registration')
    print('#2 Delete Vehicle from registration')
    print('#3 View Current registration')
    print('#4 Update Vehicle in registration')
    print('#5 Export Current registration')
    print('#6 Quit')
    userInput=input('Please choose from one of the above options: ') 
    if userInput=="1": 
        registration.addVehicle()
    elif userInput=='2':
        if len(registration.vehicles) < 1:
            print('Sorry there are no vehicles currently in registration')
            continue
        registration.viewregistration()
        item = int(input('Please enter the number associated with the vehicle to be removed: '))
        if item - 1  > len(registration.vehicles):
            print('This is an invalid number')
        else:
            registration.vehicles.remove(registration.vehicles[item - 1])
            print ()
            print('This vehicle has been removed')
    elif userInput == '3':
        if len(registration.vehicles) < 1:
            print('Sorry there are no vehicles currently in registration')
            continue
        registration.viewregistration()
    elif userInput == '4':
        if len(registration.vehicles) < 1:
            print('Sorry there are no vehicles currently in registration')
            continue
        registration.viewregistration()
        item = int(input('Please enter the number associated with the vehicle to be updated: '))
        if item - 1  > len(registration.vehicles):
            print('This is an invalid number')
        else:
            automobile = Automobile()
            if automobile.addVehicle() == True :
                registration.vehicles.remove(registration.vehicles[item - 1])
                registration.vehicles.insert(item - 1, automobile)
                print ()
                print('This vehicle has been updated')
    elif userInput == '5':
        
        if len(registration.vehicles) < 1:
            print('Sorry there are no vehicles currently in registration')
            continue
        f = open('vehicle_registration.txt', 'w')
        f.write('\t'.join(['Make', 'Model','Year', 'Color', 'Mileage']))
        f.write('\n')
        for vechicle in registration.vehicles:
            f.write('%s\n' %vechicle)
        f.close()
        print('The vehicle registration has been exported to a file')
    elif userInput == '6':
        print('Goodbye')
        break
    else:
        print('This is an invalid input. Please try again.')
```
